```python
import torch
from torch.autograd import Variable
import torch.nn.functional as F
```

```python
x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0], [4.0]]))
y_data = Variable(torch.Tensor([[0.], [0.], [1.], [1.]]))
```

```python
class Model(torch.nn.Module):
    
    def __init__(self):
        """
        to instantiate nn.Linear
        """
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1, 1)
        
    def forward(self, x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred
    
    
```

```python
# instantiation

logistic = Model()
```

```python
# Construct our loss function and an Optimizer
criterion = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(logistic.parameters(), lr=0.01)
```

```python
# Training our logistic model

for epoch in range(1000):
    
    # calculate prediction
    y_pred = logistic(x_data)
    
    # calculate loss
    loss = criterion(y_pred, y_data)
    
    print(epoch, loss.data[0])
    
    # optimization i.e. pytorch rythm forward, backward, update
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# prediction
# After training
hour_var = Variable(torch.Tensor([[1.0]]))
print("predict 1 hour ", 1.0, logistic(hour_var).data[0][0] > 0.5)
hour_var = Variable(torch.Tensor([[7.0]]))
print("predict 7 hours", 7.0, logistic(hour_var).data[0][0] > 0.5)
```
